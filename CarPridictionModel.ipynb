## Lets clean data and remove inconsistencies

```python
#Clean year by removing non numeric values
car = car[car['year'].str.isnumeric()]
car
```

```python
#Rows where year has no numeric values are being cleaned. Now data has 842 columns intead of 892
```

```python
#correcting data type
car['year']=car['year'].astype(int)
```

```python
car['year'].dtype
```

```python
car['Price'].unique
```

```python
car['Price'].str.isnumeric()
```

```python
car = car[car['Price']!= "Ask For Price"]
```

```python
car['Price']
```

```python
# Remove commas from price column and convert into int type
car['Price'] = car['Price'].str.replace(',', '').astype(int)
```

```python
# Remove kms with int in kms_driven column
car['kms_driven'] = car['kms_driven'].str.split(' ').str.get(0)
```

```python
#Remove commas
car['kms_driven'] = car['kms_driven'].str.replace(',', '')
```

```python
#Droping rows of non-int values
car = car[car['kms_driven'].str.isnumeric()]
```

```python
car['kms_driven'] = car['kms_driven'].astype(int)
```

```python
car['kms_driven'].dtype
```

```python
car.info()
```

```python
#remove NaN value from column fuel_type
car = car[~car['fuel_type'].isna()]
```

```python
# name column remove to three words and setting type as str
car['name'] = car['name'].str.split().str.slice(0,3).str.join(' ')
```

```python
car['name'] = car['name'].astype(str)
```

```python
car.describe()
```

```python
# Price column seems to have outlier. Removing skewed values
car = car[car['Price']<60e5]
```

```python
car
```

```python
# Reseting index after cleaning data
car.reset_index(drop = True)
```

## store cleaned data in csv

```python
car.to_csv('cleaned_car_data.csv')
```

## Model

```python
x = car[['name', 'company', 'year', 'kms_driven', 'fuel_type']]
y = car['Price']
```

```python
# Train Test Split
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train,y_test = train_test_split(x,y, test_size = 0.2)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
```

```python
#Creating an OneHotEncoder object to contain all the possible categories
```

```python
ohe = OneHotEncoder()
ohe.fit(x[['name', 'company', 'fuel_type']])
```

```python
ohe.categories_
```

```python
#Creating a column transformer to transform categorical columns
```

```python
column_trans = make_column_transformer((OneHotEncoder(categories = ohe.categories_), ['name', 'company', 'fuel_type']), remainder = 'passthrough')
```

```python
#Linear Regression Model
```

```python
lr = LinearRegression()
```

```python
#Making a pipeline
```

```python
pipe = make_pipeline(column_trans, lr)
```

```python
#Fitting the model
```

```python
pipe.fit(x_train, y_train)
```

```python
y_predic = pipe.predict(x_test)
```

```python
#Checking R2 Score
```

```python
r2_score(y_test, y_predic)
```

```python
#Finding the model with a random state of TrainTestSplit where the model has high r2_score
```

```python
scores = []
for i in range(1000):
    x_train, x_test, y_train,y_test = train_test_split(x,y, test_size = 0.2, random_state = i)
    lr = LinearRegression()
    pipe = make_pipeline(column_trans, lr)
    pipe.fit(x_train, y_train)
    y_predic = pipe.predict(x_test)
    scores.append(r2_score(y_test, y_predic))
```

```python
np.argmax(scores)
```

```python
scores[np.argmax(scores)]
```

```python
#The best model is found at a certain random state
```

```python
x_train, x_test, y_train,y_test = train_test_split(x,y, test_size = 0.2, random_state = np.argmax(scores))
lr = LinearRegression()
pipe = make_pipeline(column_trans, lr)
pipe.fit(x_train, y_train)
y_predic = pipe.predict(x_test)
r2_score(y_test, y_predic)
```
